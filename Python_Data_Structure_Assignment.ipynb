#Data Types and Structures Questions

1. What are data structures, and why are they important?
- Data structures are way of organizing and storing data so that it can be accessed and manipulated easily.They are important for efficiency, making code more organized performance of algorithms and readability.

2. Explain the difference between mutable and immutable data types with examples.
- Mutable data types can be changed after they are created.Examples:List,Dictionaries and Sets.
Immutable data types cannnot be changed after they created.Examples:Strings,Tuples and Numbers(Integers,Floats).

3. What are the main differences between lists and tuples in Python?
- Lists are mutable, while tuples are immutable.
- Lists are defined using square brackets[],while tuples are defined using parenthesis().
- Lists are used for collection of items that might change,while tuples are used for fixed collection of items.

4. Describe how dictionaries store data.
- Dictionaries store data as a key-value pairs where each unique maps to a specific value.

5. Why might you use a set instead of a list in Python?
- We might use as a set when we need to store a collection of unique items and perform operation like union,intersection ,or difference,or when efficient membership testing is required.

6. What is a string in Python, and how is it different from a list?
- A string in Python is an immutable sequence of characters,while a list is a mutable sequence of fixed objects.

7. How do tuples ensure data integrity in Python?
- They ensure data integrity because their immutability prevents accidental modification of their elements after creation.

8. What is a hash table, and how does it relate to dictionaries in Python?
- It is a data structure that implements an associative array abstract data type,mapping keys to values using hash function Python dictionaries are implemented using hash tables.

9. Can lists contain different data types in Python?
- Yes lists in Python can contain elemets of different data types.

10. Explain why strings are immutable in Python.
- Strings are immutable in Python because once they created their content cannot be changed

11. What advantages do dictionaries offer over lists for certain tasks?
- Dictionaries offer faster data retrieval by key, efficient storge of related data as key-value pairs,and clearer organization when data has meaningful lables.

12. Describe a scenario where using a tuple would be preferable over a list.
- It is preferable when representation a fixed collection of related items that should not change data such as RGB colour values(red,green,blue).

13. How do sets handle duplicate values in Python?
- Set automatically eliminate duplicate values,storing only unique elements.

14. How does the “in” keyword work differently for lists and dictionaries?
- For lists,the 'in' keyword checks for the presence of a value among the lists elements,while for dictionaries,it checks for the presence of a key.

15. Can you modify the elements of a tuple? Explain why or why not.
- No we cannot modify the the elements of a tuple because tuples are immutable.

16. What is a nested dictionary, and give an example of its use case?
- A nested dictionary is a dictionary within another dictionary. It is useful for storing hierarchical data, such as a directory of users where each user has their personal data stored as a dictionary of properties.

17. Describe the time complexity of accessing elements in a dictionary.
- The average case time complexity for accessing elements in a dictionary is O (1) (constant time).This efficiency makes dictionaries ideal for scenarios that require frequent and rapid data retreival.

18. In what situations are lists preferred over dictionaries?
- Lists are preferred when the order of elements is important,when you need to store a collection of items that might contain duplicates,or when you primarily access elements by their numerical index.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictinaries are considered unordered because in many programming languages(especially older versions of Python before 3.7), they do not guarantee the order of elements(key-value pairs ) when they are stored or iterated over.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- Data in a list is typically retreived by its numerical index(position), while data in a dictionary is retreived by its asscosiated key.
